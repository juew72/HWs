```python
import pandas as pd
import time
import datetime as dt
```

```python
df = pd.read_csv('ustsv_data_thc.csv')
df
```

```python
#remove T in df['event_time']
df['event_time'].replace(regex=True,inplace=True,to_replace=r'T',value=r' ')
df
```

```python
df['datetime'] = pd.to_datetime(df['event_time'])
#df = df.set_index('datetime')
df.drop(['event_time'], axis=1, inplace=True)
#df.head()
```

```python
df['sec'] = (df['datetime'] - dt.datetime(2019,12,17)).dt.total_seconds()
df.head()
```

```python
def correct_second(n):
    #corrected_time = []
    corrected_second = [df.sec.iloc[0]]
    for i in range(0, len(df['sec'])-1):
        if df.sec.iloc[i] == df.sec.iloc[i+1]:
            df.sec.iloc[i+1] += 1
        elif df.sec.iloc[i] > df.sec.iloc[i+1]:
            df.sec.iloc[i] += 1
            df.sec.iloc[i+1] += 1
            df.sec.iloc[i], df.sec.iloc[i+1] = df.sec.iloc[i+1], df.sec.iloc[i]
            if df.sec.iloc[i] == df.sec.iloc[i-1]:
                df.sec.iloc[i] += 1
        corrected_second.append(df.sec.iloc[i+1])
    #print(df.sec.iloc[i+1])
    
    #reference_date1 = dt.datetime(2019, 12, 17, 0, 0, 0)
    #time = dt.datetime.fromtimestamp(df.sec.iloc[i+1] + float(dt.datetime.strptime(str(reference_date1), 
    #                                                                        "%Y-%m-%d %H:%M:%S").strftime('%s')))
    #corrected_time.append(time)
    #print(time)
    
    df['corrected_second'] = corrected_second
    
    
    
    return n
```

```python
def correct_time(n):
    corrected_time = []
    for i in range(0,len(df.corrected_second)):
        reference_date1 = dt.datetime(2019, 12, 17, 0, 0, 0)
        time = dt.datetime.fromtimestamp(i + float(dt.datetime.strptime(str(reference_date1),"%Y-%m-%d %H:%M:%S").strftime('%s')))
        corrected_time.append(time)
    df['corrected_time'] = corrected_time
    return n
```
